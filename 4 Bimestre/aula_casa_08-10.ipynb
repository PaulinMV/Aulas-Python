```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Dados: Exemplo
areas = np.array([50, 60, 80, 100, 120]).reshape(-1, 1) # Variável independente (x)
precos = np.array([150, 180, 240, 300, 360]) # Variável dependente (y)
```

```python
# Criação do modelo de regressão linear
modelo = LinearRegression()
modelo.fit(areas, precos)
```

```python
# Coeficientes
m = modelo.coef_[0]
b = modelo.intercept_
```

```python
# Previsão para uma casa de 90m²
area_nova = np.array([[90]])
preco_previsto = modelo.predict(area_nova)
```

```python
# Visualizando o resultado
plt.scatter(areas, precos, color='magenta', label='Dados Reais')
plt.plot(areas, modelo.predict(areas), color='black', label=f'Reta de Regressão\n (y = {m:.2f}x + {b:.2f})')
plt.scatter(area_nova, preco_previsto, color='blue', label=f'Previsão para 90m²: R$ {preco_previsto[0]:.2f}')
plt.title('Regressão Linear Simples: Área vs Preço de Casas')
plt.xlabel('Àrea (m²)')
plt.ylabel('Preço (R$ mil)')
plt.legend()
plt.show()
```
