```python
#importação da biblioteca match
import math
```

```python
#Construção da equação de segundo grau de bhaskara
def delta (a, b, c):
  delta = b ** 2 - 4*a*c
  if (delta > 0):
    x1 = (-b+math.sqrt(delta))/(2*a)
    x2 = (-b-math.sqrt(delta))/(2*a)
    return x1, x2
  elif (delta == 0):
    x1 = -b/(2*a)
    return x1
  else:
    return ("Não existem raizes reais")
a = float(input("Digite o valor de a: "))
b = float(input("Digite o valor de b: "))
c = float(input("Digite o valor de c: "))

resultado = delta(a,b,c)
print(f"o valor de delta é: {resultado}")
```
