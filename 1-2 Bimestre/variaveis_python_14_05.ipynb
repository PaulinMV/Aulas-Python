Variaveis atribuidas

```python
largura = 15
altura = 23

area = largura * altura

print(area)
```

```python
idade_1 = 20
idade_2 = 60
idade_1 + idade_2
```

```python
idade_soma = idade_1 + idade_2
print(idade_soma)
```

```python
idade_subtrac = idade_2 - idade_1
print(idade_subtrac)
```

```python
idade_multiplic = idade_1 * idade_2
print(idade_multiplic)
```

```python
idade_div = idade_1 / idade_2
print(idade_div)
```

```python
print(idade_soma)
print(idade_subtrac)
print(idade_multiplic)
print(idade_div)
```
