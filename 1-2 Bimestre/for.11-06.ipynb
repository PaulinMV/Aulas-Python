Loop FOR

```python
#For que vai até 10
for x in range(10):
  print(x)
```

```python
#For que vai do 10 e termina no 20, detalhe: ele não finaliza antes de incluir o 20
for x in range(10,20):
  print(x)
```

```python
#Parametro pra começo, fim e um i++ que ele faz de 3 em 3 nesse caso
for x in range (2, 30, 3):
  print(x)
```

```python
#For com parametro de array
lojas = ['são paulo', 'rio de janeiro', 'parana']

for loja in lojas:
  print(loja)

print('Tudo certo!')
```

```python
#For que imprime valor de i e o valor do array loja de acordo com o i
for i in range(3):
  print(i)
  print(lojas[i])
```

```python
#For que irá imprimir x do tamanho da palavra
for x in 'Paralelepipedo':
  print(x)
```

```python
#For que faz uma varredura com parametro de parada, se encontrar a variavel banana, ele para
frutas = ['maça', 'banana', 'morango']
for x in frutas:
  print(x)
  if x == 'banana':
    break
```

```python
#len faz a leitura de quantos itens tem no array frutas
len(frutas)
```

```python
#Esse for com o print por fora, vai mostrar so o resultado da parada, nesse caso a variavel 'banana'
frutas = ['maça', 'banana', 'morango']
for x in frutas:
  if x == 'banana':
    break
print(x)
```

```python
#Esse for ao contrario do outro que quando encontra a variavel ele para, esse imprimira somente a variavel que ele encontrar com o metodo continue
frutas = ['maça', 'banana', 'morango']
for x in frutas:
  if x == 'maça':
    continue
print(x)
```

```python
#Um for dentro do outro
cores = ['vermelho', 'verde', 'azul']
carros = ['corolla', 'civic', 'azera']

for x in cores:
  for y in carros:
    print (x, y)
```
