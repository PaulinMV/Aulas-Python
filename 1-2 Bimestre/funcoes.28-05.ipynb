# Funções Python

```python
def funcao(): #declarando uma função SEM parâmetros
  print('este é um teste de função!')
  print('hello world')

funcao() #testando a função
print(funcao()) #imprimindo a função
```

```python
def nacional(pais): #declarando uma função COM parâmetros
  print("Eu venho do(a) " + pais)

nacional("Brasil")
nacional("França")
nacional("Chile")
nacional("Japão")
nacional("China")
```

```python
#testando com um input como forma de argumento
def nomePais(name, pais):
  print("Olá, " + name)
  print("Você é do(a): " + pais)

name = input("insira seu nome: ")
pais = input("Insira seu País de origem: ")
print()
print (nomePais(name, pais))
```

```python
def telegrama(mensagem): #função COM argumento
  print(mensagem)

telegrama("Hello World")
```

```python
#REPLACE

def saudacao(msg = "olá", nome = "usuario"):
  msg = msg.replace('o', '0')
  nome = nome.replace('a', '4')
  print(msg, nome)

saudacao('olá,', 'Paulo')
saudacao ('Hello', 'World')
```

```python
#RETURN
def saudacao(msg = "Olá", nome = "Paulo"):
  return f'{msg} {nome}, seja bem vindo!'

variavel = saudacao()
print (variavel)
```

```python
def soma (x, y):
  return x + y

def subtrac (x,y):
  return x - y

def multiplic (x,y):
  return x * y

def div (x,y):
  return x / y

print (soma (2, 2))
print (subtrac (24, 12))
print (multiplic (5, 5))
print (div (54, 6))
```
