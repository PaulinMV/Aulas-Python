Formula de Bhaskara

```python
def bhaskara (a, b,c):
  delta = b ** 2 - 4 * a * c
  if delta < 0:
    return "A equação não possui raizes reais"
  elif delta == 0:
    x = -b / (2 * a)
    return f"A equação possui uma raiz real: x = {x}."
  else:
    x1 = (-b + delta ** 0.5) / (2 * a)
    x2 = (b + delta ** 0.5) / (2 * a)
    return f"A equação possui duas raízes reais: x1 = {x1} e x2 = {x2}"
```
