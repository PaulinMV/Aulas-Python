# **TESTES DE ENTRADA INVÁLIDA - 13/08**
Neste exemplo, vamos verificar se uma função lida corretamente com entradas inválidas e retorna os resultados ou mensagens de erro apropriados.

```python
#Criando o arquivo math_utilities.py
%%writefile math_utilities.py
import math

class ValorInvalidoError(Exception):
  """Exceção personalizada para valores inválidos."""
  pass
```

```python
def calcular_raiz_quadrada(valor):
  """
  Calcula a raiz quadrada de um número, levantando uma exceção se o valor for inválido.

  Parâmetros:
    valor (float ou int): O número para calcular a raiz quadrada.

  Retorna:
    float: A raiz quadrada do valor.

  Lança:
    ValorInvalidoError: Se o valor for negativo ou não for um número.
  """

  if not isinstance(valor, (int, float)):
    raise ValorInvalidoError("O valor deve ser um número.")
  if valor < 0:
    raise ValorInvalidoError("Não é possível calcular a raiz quadrada de um número negativo.")
  return math.sqrt(valor)
```

```python
#Criando o aarquivo de teste test_math_utilities.py
%%writefile test_math_utilities.py
import pytest
from math_utilities import calcular_raiz_quadrada, ValorInvalidoError

def test_calcular_raiz_quadrada():
  #Teste com valores válidos
  assert calcular_raiz_quadrada(4) == 2
  assert calcular_raiz_quadrada(9) == 3
  assert calcular_raiz_quadrada(0) == 0
  assert pytest.approx(calcular_raiz_quadrada(2)) = 1.4142135623730951  # Comparação de floats com aproximação

  #Teste com valor negativo
  with pytest.raises(ValorInvalidoError) as excinfo_negativo:
    calcular_raiz_quadrada(-4)
  assert str(excinfo_negativo.value) == "Não é possivel calcular a raiz quadrada de um número negativo."

  #Teste com entrada não númerica
  with pytest.raises(ValorInvalidoError) as excinfo_tipo:
    calcular_raiz_quadrada("abc")
  assert str(excinfo_tipo.value) == "O valor deve ser um número."

  with pytest.raises(ValorInvalidoError) as excinfo_tipo:
    calcular_raiz_quadrada([1,2,3])
  assert str(excinfo_tipo.value) == "O valor deve ser um número."
```
