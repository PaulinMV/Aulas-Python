# **TESTE UNITARIO 2**

PASSO 1: Criar o arquivo my_functions.py com as funções a serem testadas

```python
%%writefile my_functions.py
def add(a, b):
  return a + b

def subtract(a, b):
  return a - b

def multiply(a, b):
  return a * b
```

PASSO 2: Criar o arquivo de teste test_my_functions.py

```python
%%writefile test_my_functions.py
import pytest
from my_functions import add, subtract, multiply

def test_add():
  assert add(4, 5) == 9
  assert add(-2, 3) == 1

def test_subtract():
  assert subtract(20, 10) == 10
  assert subtract(15, 16) == -1

def test_multiply():
  assert multiply(2, 2) == 4
  assert multiply(5, 9) == 45
```

PASSO 3: Testar através do PYTEST

```python
!pytest test_my_functions.py
```
