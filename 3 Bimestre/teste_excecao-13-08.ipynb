# **TESTE DE EXCEÇÃO - 13/08**
Objetivo: Garantir que o código manipule corretamente situações de erro e lance exceções apropriadas.
Exemplo: Testar se uma função lança uma exceção quando recebe um valor inválido.

```python
#Criando o arquivo math_operations.py
%%writefile math_operations.py
class DivisaoPorZeroError(Exception):
  """Exceção personalizada para divisão por zero."""
  pass

def dividir(numerador, denominador):
  """Divide o numerador pelo denominador e lança uma exeção se o denominador for zero."""
  if denominador == 0:
    raise DivisaoPorZeroError("Não é possível dividir por zero.")
  return numerador / denominador
```

```python
#Criando o arquivo de teste test_math_operations.py
%%writefile test_math_operations.py
import pytest
from math_operations import dividir, DivisaoPorZeroError

def test_dividir():
  #Valores válidos para teste
  assert dividir(10,2) == 5
  assert dividir(-10,2) == -5
  assert dividir(0,1) == 0

  #Valor para teste de divisão por zero
  with pytest.raises(DivisaoPorZeroError) as excinfo:
    dividir(10,0)
  assert str(excinfo.value) == "Não é possível dividir por zero."
```
