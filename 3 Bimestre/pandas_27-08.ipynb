```python
# Importação da bibilioteca
import pandas as pd
```

```python
# Series:
# Uma Series é uma estrutura de dados unidimensional,
# semelhante a uma lista ou um array, que pode conter qualquer
# tipo de dados (inteiros, strings, floats, etc.).

serie_inteiros = pd.Series([10, 20, 30, 40, 50])
print(serie_inteiros)
```

```python
serie_inteiros
```

```python
# Serie com Strings
serie_strings = pd.Series(['maça', 'uva', 'pera', 'damasco', 'banana'])
print(serie_strings)
```

```python
serie_strings
```

```python
# Serie com valores Booleanos
serie_booleanos = pd.Series([True, False, True, False, True])
print(serie_booleanos)
```

```python
serie_booleanos
```

```python
# Série com numeros decimais
serie_floats = pd.Series([1.5, 2.75, 3.25, 4.0, 5.5])
print(serie_floats)
```

```python
serie_floats
```

```python
# Serie com datas
serie_datas = pd.Series(pd.date_range('2024-03-06', periods=5))
print(serie_datas)
```

```python
serie_datas
```

# DataFrame

Um DataFrame é uma estrutura de dados bidimensional, como uma tabela, que é usada para armazenar e manipular dados em formato tabular em Python, especialmente na biblioteca Pandas. Ele é composto de linhas e colunas, onde cada coluna pode conter um tipo de dado diferente (como números inteiros, floats, strings, etc.).

```python
df_4_colunas = pd.DataFrame({
    'Nome': ['Ana', 'Bruno', 'Carlos'],
    'Idade': ['23', '35', '45'],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte'],
    'Profissão': ['Engenheira', 'Médico', 'Professor']
})
df_4_colunas
```

```python
df_5_colunas = pd.DataFrame ({
    'Produto': ['Notebook', 'Smartphone', 'Tablet'],
    'Preço': [3000, 1500, 1000],
    'Quantidade': [10, 20, 30],
    'Fornecedor': ['Empresa A', 'Empresa B', 'Empresa C'],
    'Garantia': ['2 anos', '1 ano', '1 ano']
})
df_5_colunas
```

```python
df_6_colunas = pd.DataFrame({
    'Aluno': ['João', 'Maria', 'Pedro'],
    'Matemática': [8.5, 7.0, 9.0],
    'Português': [7.5, 8.0, 6.0],
    'História': [9.0, 6.5, 8.0],
    'Geografia': [8.0, 7.5, 9.5],
    'Ciências': [7.5, 9.0, 8.5]
})
df_6_colunas
```

```python
df_7_colunas = pd.DataFrame ({
    'Data': pd.date_range(start='2024-01-01', periods=3, freq='D'),
    'Vendedor': ['Carlos', 'Fernanda', 'Marcos'],
    'Produto': ['Camiseta', 'Calça', 'Sapato'],
    'Quantidade': [50, 40, 30],
    'Preço Unitário': [20, 50, 100],
    'Desconto': [5, 10, 15],
    'Total': [950, 1800, 2550]
})
df_7_colunas
```

```python
df_8_colunas = pd.DataFrame({
    'Empresa': ['Empresa X', 'Empresa Y', 'Empresa Z'],
    'Receita': [500000, 750000, 1000000],
    'Custo': [300000, 450000, 600000],
    'Lucro': [200000, 300000, 400000],
    'Funcionários': [50, 75, 100],
    'Localização': ['SP', 'RJ', 'MG'],
    'Segmento': ['Tecnologia', 'Varejo', 'Indústria'],
    'Fundação': [2000, 1995, 1985]
})
df_8_colunas
```
