```python
# Criando o arquvio que contem as funções a serem executadas
%%writefile string_functions.py

# Função que retorna o numero de caracteres da variavel 'texto':
def contar_caracteres(texto):
  return len(texto)

# Função que retorna quantas palavras estão presentes na variavel 'texto'
def contar_palavras(texto):
    return len(texto.split())

# Função que retorna a string 'texto' invertida
def inverter_string(texto):
    return texto[::-1]

# Função que retorna True se a string for um palíndromo
# (ignorando maiúsculas/minúsculas e espaços), False caso contrário.
def eh_palindromo(texto):
    texto = texto.replace(" ", "").lower()
    return texto == texto[::-1]
```

```python
# Criando o arquivo que contem os testes:
%%writefile test_string_operations.py

import pytest
from string_functions import contar_caracteres, contar_palavras, inverter_string, eh_palindromo

# Função que testa a função importada: contar_caracteres
def test_contar_caracteres():
  assert contar_caracteres('Python') == 6
  assert contar_caracteres('') == 0
  assert contar_caracteres('Paralelepipedo') == 14

# Função que testa a função importada: contar_palavras
def test_contar_palavras():
  assert contar_palavras('Hello World') == 2
  assert contar_palavras('Olá meu nome é Paulo') == 5
  assert contar_palavras('') == 0

# Função que testa a função importada: inverter_string
def test_inverter_string():
  assert inverter_string('Onibus') == 'subinO'
  assert inverter_string('Garrafa') == 'afarraG'
  assert inverter_string('') == ''

# Função que testa a função importada: eh_palindromo
def test_eh_palindromo():
  assert eh_palindromo('ArarA') == True
  assert eh_palindromo('subi no onibus') == True
  assert eh_palindromo('rua') == False
```
