1 - Criando o arquvivo com as funções para teste:

```python
%%writefile string_functions.py
#Função que retorna a VARIAVEL palavra apenas em letras maiusculas:
def para_maiusculas(palavra):
  return palavra.upper()

#Função que retorna o numero de caracteres da VARIAVEL palavra:
def contar_caracteres(palavra):
  return len(palavra)

#Função que retorna apenas as letras da VARIAVEL palavra:
def apenas_letras(palavra):
  return palavra.isalpha()
```

2 - Criando o arquivo com as funções que irão fazer o teste:

```python
%%writefile test_string_functions.py
import pytest
from string_functions import para_maiusculas, contar_caracteres, apenas_letras

def test_para_maiusculas():
  assert para_maiusculas('python') == 'PYTHON'
  assert para_maiusculas('PyThOn') == 'PYTHON'
  assert para_maiusculas('123') == '123'

def test_contar_caracteres():
  assert contar_caracteres('python') == 6
  assert contar_caracteres('') == 0
  assert contar_caracteres('OpenAI') == 6

def test_apenas_letras():
  assert apenas_letras('python') == True
  assert apenas_letras('123') == False
  assert apenas_letras('Python3') == False
  assert apenas_letras('AI') == True
```

3 - PYTEST
